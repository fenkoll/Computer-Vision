```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from PIL import Image
from scipy import misc
from scipy import ndimage
from skimage import color
from skimage import io
```

```python
img1 = color.rgb2gray(io.imread('part1/c1.jpg'))
plt.figure(dpi=100)
plt.axis('off')
plt.imshow(img1, cmap="gray") 
```

```python
img2 = color.rgb2gray(io.imread('part1/c2.jpg'))
plt.figure(dpi=100)
plt.axis('off')
plt.imshow(img2, cmap="gray") 
```

```python
img1_lowPass=ndimage.gaussian_filter(img1, sigma=1, order=0)#, output=None, mode='reflect', cval=0.0, truncate=4.0
plt.figure(dpi=100)
plt.axis('off')
plt.imshow(img1_lowPass, cmap="gray") 
```

```python
img2_lowPass=ndimage.gaussian_filter(img2, sigma=10)#, order=0, output=None, mode='reflect', cval=0.0, truncate=4.0
img2_highPass=img2-img2_lowPass
plt.figure(dpi=100)
plt.axis('off')
plt.imshow(img2_highPass, cmap="gray") 
```

```python
hybrid_image1=img1_lowPass+img2_highPass
plt.figure(dpi=200)
plt.axis('off')
plt.imshow(hybrid_image1, cmap="gray") 
```

```python
hybrid_image1=img1_lowPass+img2_highPass
plt.figure(dpi=40)
plt.axis('off')
plt.imshow(hybrid_image1, cmap="gray") 
```
